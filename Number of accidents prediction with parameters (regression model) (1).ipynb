```python
import matplotlib.pyplot as plt
print('                                    LINEARITY ANALYSIS FOR TWO WHEELERS')
plt.plot(df['TWO_WHEELER'], df['OVERSPEEDING'], color='red')
plt.title('TWO_WHEELER Vs OVERSPEEDING', fontsize=14)
plt.xlabel('TWO_WHEELER', fontsize=14)
plt.ylabel('OVERSPEEDING', fontsize=14)
plt.grid(True)
plt.show()

plt.plot(df['TWO_WHEELER'], df['OVERTAKING'], color='red')
plt.title('TWO_WHEELER Vs OVERTAKING', fontsize=14)
plt.xlabel('TWO_WHEELER', fontsize=14)
plt.ylabel('OVERTAKING', fontsize=14)
plt.grid(True)
plt.show()

plt.plot(df['TWO_WHEELER'], df['WRONG_ROUTE'], color='red')
plt.title('TWO_WHEELER Vs WRONG_ROUTE', fontsize=14)
plt.xlabel('TWO_WHEELER', fontsize=14)
plt.ylabel('WRONG_ROUTE', fontsize=14)
plt.grid(True)
plt.show()

plt.plot(df['TWO_WHEELER'], df['RASH_DRIVING'], color='red')
plt.title('TWO_WHEELER Vs RASH_DRIVING', fontsize=14)
plt.xlabel('TWO_WHEELER', fontsize=14)
plt.ylabel('RASH_DRIVING', fontsize=14)
plt.grid(True)
plt.show()

plt.plot(df['TWO_WHEELER'], df['DISTRACTED_DRIVING'], color='red')
plt.title('TWO_WHEELER Vs DISTRACTED_DRIVING', fontsize=14)
plt.xlabel('TWO_WHEELER', fontsize=14)
plt.ylabel('DISTRACTED_DRIVING', fontsize=14)
plt.grid(True)
plt.show()

plt.plot(df['TWO_WHEELER'], df['DRUNKEN_DRIVING'], color='red')
plt.title('TWO_WHEELER Vs DRUNKEN_DRIVING', fontsize=14)
plt.xlabel('TWO_WHEELER', fontsize=14)
plt.ylabel('DRUNKEN_DRIVING', fontsize=14)
plt.grid(True)
plt.show()

plt.plot(df['TWO_WHEELER'], df['NEGLIGENCE'], color='red')
plt.title('TWO_WHEELER Vs NEGLIGENCE', fontsize=14)
plt.xlabel('TWO_WHEELER', fontsize=14)
plt.ylabel('NEGLIGENCE', fontsize=14)
plt.grid(True)
plt.show()


plt.plot(df['TWO_WHEELER'], df['FAILURE_OF_VEHICLE'], color='red')
plt.title('TWO_WHEELER Vs FAILURE_OF_VEHICLE', fontsize=14)
plt.xlabel('TWO_WHEELER', fontsize=14)
plt.ylabel('FAILURE_OF_VEHICLE', fontsize=14)
plt.grid(True)
plt.show()
```

```python
X = df[['OVERSPEEDING','OVERTAKING','WRONG_ROUTE','RASH_DRIVING','DISTRACTED_DRIVING','DRUNKEN_DRIVING','NEGLIGENCE','FAILURE_OF_VEHICLE']]
Y = df['TWO_WHEELER']
from sklearn import linear_model
regr = linear_model.LinearRegression()
regr.fit(X, Y)

print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)
```

```python
print("TWO_WHEELER_ACCIDENT_PREDICITON = ",regr.intercept_," + (",regr.coef_[0],")*(Overspeeding)","                                                                   + (",regr.coef_[1],")*(Overtaking)"," + (",regr.coef_[2],")*(Wrong route)","                                             + (",regr.coef_[3],")*(Rash Driving)"," + (",regr.coef_[4],")*(DISTRACTED DRIVING)","                                     + (",regr.coef_[5],")*(DRUNKEN_DRIVING)"," + (",regr.coef_[6],")*(NEGLIGENCE)","                                              + (",regr.coef_[7],")*(FAILURE_OF_VEHICLE)")
```
